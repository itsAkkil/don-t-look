```python
y = df_new['attack']
X = df_new[['rcvTime', 'pos_0', 'pos_1', 'pos_noise_0',
       'pos_noise_1', 'spd_0', 'spd_1', 'spd_noise_0', 'spd_noise_1', 'acl_0',
       'acl_1', 'acl_noise_0', 'acl_noise_1', 'hed_0', 'hed_1', 'hed_noise_0',
       'hed_noise_1']]
```

```python
# step 4: train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_pred,y_test)
print("Logistic Regression")
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

**KNeighborsClassifier**

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
knc.fit(X_train,y_train)
y_pred_k=knc.predict(X_test)
```

```python
score=accuracy_score(y_pred_k,y_test)
print("Kneighbour classifier")
print(score)
print(classification_report(y_pred_k,y_test))
print(confusion_matrix(y_pred_k,y_test))
```

**Support vector classification**

```python
from sklearn.svm import SVC
svc=KNeighborsClassifier()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("support vector")
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc=KNeighborsClassifier()
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print('Decision Tree Classifier')
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=KNeighborsClassifier()
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print('Random Forest Classifier')
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
